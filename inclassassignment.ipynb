```python
import random

play = input("Do you want to play? (y/n): ").lower()

while play == "y":
  p_move = input("Choose between rock, paper, scissors: ").lower()
  c_move = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player move: {p_move}")
  print(f"Computer move: {c_move}")

  if p_move == c_move:
    print("Draw")
  elif (
      (p_move == "rock" and c_move == "scissors") or
      (p_move == "scissors" and c_move == "paper") or
      (p_move == "paper" and c_move == "rock")
  ):
    print("Player wins!")
  else:
    print("Computer wins!")

  play = input("Do you want to play again? (y/n): ")
```

```python
def roi(revenue, cost):
  assert cost != 0
  assert isinstance(revenue, (int, float))
  assert isinstance(cost, (int, float))
  return (revenue - cost) / cost
```

```python
def loan_payment(principal, annual_rate, months):
  assert isinstance(principal, (int, float))
  assert isinstance(annual_rate, (int, float))
  assert isinstance(months, int)

  if principal <= 0 or annual_rate <= 0 or months <= 0:
    raise valueError("Invalid input values")

  try:
    r = annual_rate / 12 / 100
    if r == 0:
      pmt = principal / months
    else:
      pmt = r * principal / (1 - (1 + r) ** (-months))
  except Exception as e:
    raise Exception(f"Error: {e}")

  return pmt
```

```python
class Greeting:
  def __init__(self, first, last):
    self.first = first
    self.last = last

  def display(self, formal=False):
    if formal:
      print(f"{self.last}, {self.first}")
    else:
      print(f"{self.first}")

  def __str__(self):
    return "Class Greeting \nReturns formal or informal greetings"

s = Greeting("John", "Doe")
print(s)
s.display()
s.display(formal=True)
```

```python
class Student:
  def __init__(self, name, number):
    self.__name = name
    self.number = number
    self.__courses = []
  def get_courses(self):
    return self.__courses
  def get_name(self):
    return self.__name
  def set_name(self, newName):
    self.__name = newName
  def __len__(self):
    return len(self.__courses)
  def enroll(self, course):
    if course in self.__courses:
      print(f"Student {self.__name} has already enrolled in course {course}.")
    else:
      self.__courses.append(course)
```

```python
s = Student("Jane Doe", 12345)
print("Student name:", s.get_name())
s.set_name("Jane B. Doe")
print("Student name:", s.get_name())
print("Courses enrolled in:", s.get_courses())
s.enroll('BTE320')
s.enroll('BTE320')
print(f"Student {s.get_name()} has enrolled in {len(s)} course(s): {s.get_courses()}")
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def distance(self, other):
    dx = (self.x - other.x) ** 2
    dy = (self.y - other.y) ** 2
    dist = (dx + dy) ** 0.5
    return dist

class Rectangle:
  def __init__(self, x1, y1, x2, y2, x3, y3, x4, y4):
    self.p1 = Point(x1, y1) #upper left
    self.p2 = Point(x2, y2) #upper right
    self.p3 = Point(x3, y3) #lower left
    self.p4 = Point(x4, y4) #lower right
    self.width = self.p1.distance(self.p2)
    self.height = self.p1.distance(self.p3)
  def area(self):
    return self.width * self.height
  def perimeter(self):
    return
```

```python
class Employee:
  def __init__(self, name, hourly_wage, hours):
    self.name = name
    self.__hourly_wage = hourly_wage
    self.hours = hours

  def get_wage(self):
    return self.__hourly_wage

  def set_wage(self, new_wage):
    assert newWage > 0
    self.__hourly_wage = newWage

  def calculate_pay(self):
    return self.__hourly_wage * self.hours

  def __str__(self):
    return f"Employee Name: {self.name}\nHourly Charge: {self.__hourly_wage}$"

class Manager(Employee):
  def __init__(self, name, hourly_wage, hours, bonus):
    super().__init__(name, hourly_wage, hours)
    self.bonus = bonus #float between 0 - 1; e.g., 0.2 means 20% extra

  def calculate_pay(self):
    return (1 + self.bonus) * super().get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)
print(employee)
print(manager)
print()
print(f"Employee {employee.name} weekly salary: {employee.calculate_pay()}$.")
print(f"Manager {manager.name} weekly salary: {manager.calculate_pay()}$ that includes a {manager.bonus * 100}% bonus.")
```
