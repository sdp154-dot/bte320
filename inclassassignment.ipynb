```python
sales = float(input("Enter Sales ($): "))
rate = float(input("Enter Tax Rate (%): "))

cogs = sales * 0.4
gross_profit = sales - cogs
taxes = rate * gross_profit / 100
net = gross_profit - taxes

print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net}$")
```

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Choose numerical operator: ")

if op == "+":
    res = num1 + num2
elif op == "-":
    res = num1 - num2
elif op == "*":
    res = num1 * num2
elif op == "/":
        res = num1 / num2
elif op == "//":
    res = num1 // num2
elif op == "**":
    res = num1 ** num2
elif op == "%":
    res = num1 % num2
else:
    print("Invalid operation")
    res = None

print(f"Result: {res}")
```

```python
import random

play = input("Do you want to play? (y/n): ").lower()

while play == "y":
  p_move = input("Choose between rock, paper, scissors: ").lower()
  c_move = random.choice(['rock', 'paper', 'scissors'])

  print(f"Player move: {p_move}")
  print(f"Computer move: {c_move}")

  if p_move == c_move:
    print("Draw")
  elif (
      (p_move == "rock" and c_move == "scissors") or
      (p_move == "scissors" and c_move == "paper") or
      (p_move == "paper" and c_move == "rock")
  ):
    print("Player wins!")
  else:
    print("Computer wins!")

  play = input("Do you want to play again? (y/n): ")
```
